```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters

n_inputs = 2
n_outputs = 5
n_samples = 1000
learning_rate = 0.01
```

```python
data1 = np.random.multivariate_normal([1, 1], [[0.5, 0],[0, 0.5]], n_samples//2)
data2 = np.random.multivariate_normal([-1, -1], [[0.5, 0],[0, 0.5]], n_samples//2)
X = np.vstack((data1, data2))
```

```python
plt.scatter(data1[:, 0], data1[:, 1], label="Class 1")
plt.scatter(data2[:, 0], data2[:, 1], label="Class 2")
plt.title("Synthetic 2D Clusters")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
weights = np.random.randn(n_outputs, n_inputs)
```

```python
for x in X:
    activations = np.dot(weights, x)
    winner = np.argmax(activations)

    for j in range(n_inputs):
        weights[winner, j] += learning_rate * x[j] * (1 - weights[winner, j])

print("Training complete.")
```
